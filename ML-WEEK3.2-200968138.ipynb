```python
def cont_to_exp(x):
    if type(x) is str:
        x = x.replace('\\','')
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
        return x
    else:
        return x
```

```python
df["Review Text"] = df["Review Text"].apply(lambda x:cont_to_exp(x))
```

```python
df.head()
```

```python
# removing punctuations
df["Review Text"] = df["Review Text"].str.replace('[^\w\s]','')
df.head()
```

#### Q2

```python
from textblob import TextBlob
```

```python
df['polarity']=df['Review Text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df.head()
```

```python
#feature_cols = ['Age','Rating','Positive Feedback Count','polarity']
#X = df[feature_cols]
X = df['polarity']
Y = df['Recommended IND']
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=25)
#performing 80-20 split on the data
```

#### Q3 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(X_train.values.reshape(-1,1), y_train)
y_pred = classifier.predict(X_test.values.reshape(-1,1))
```

#### Q4

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
TN = confusion_matrix[0,0]
TP = confusion_matrix[1,0]
FP = confusion_matrix[0,1]
FN = confusion_matrix[1,1]
TNR = TN / (TN+FP)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("Specificity:",TNR)
```

```python
from sklearn.metrics import classification_report, roc_curve
print(classification_report(y_test, y_pred)) #this was mainly used for f1 score as the rest have been obtained above
```
